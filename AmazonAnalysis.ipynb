```python
#Change the Datatypes
amazon["discounted_price"]=amazon["discounted_price"].str.replace("â‚¹","").str.replace(",","")
amazon["discounted_price"] = pd.to_numeric(amazon["discounted_price"], errors='coerce').fillna(0).astype(int)

amazon["actual_price"]=amazon["actual_price"].str.replace("â‚¹","").str.replace(",","")
amazon["actual_price"] = pd.to_numeric(amazon["actual_price"], errors='coerce').fillna(0).astype(int)

#Check The Column After changes
print(amazon[["discounted_price","actual_price"]])
```

```python
#1.How does the discount percentage affect the rating of a product?

# Non-numeric values ko NaN mein convert karke handle karna
amazon["rating"] = pd.to_numeric(amazon["rating"], errors='coerce')

# NaN values ko fill karna (0 ya kisi aur suitable value se)
amazon["rating"] = amazon["rating"].fillna(0)

# Float se round karke integer mein convert karna
amazon["rating"] = amazon["rating"].round().astype(int)



dis_per_rating=amazon.groupby(["discount_percentage"])["rating"].mean().reset_index().sort_values(by="rating",ascending=False).head(10)
dis_per_rating
```

```python
#Plotting The Data
import matplotlib.pyplot as plt

plt.plot(dis_per_rating["discount_percentage"], dis_per_rating["rating"], marker='o')
plt.xlabel("Discount Percentage")
plt.ylabel("Average Rating")
plt.title("Effect of Discount Percentage on Product Rating")
plt.show()
```

```python
#2.Which category has the highest average rating?

#Split the Category Data
amazon["New_category"]=amazon["category"].str.split("|").str[0]
category_high_rating=amazon.groupby(["New_category"])["rating"].mean()
category_high_rating
```

```python
#Plotting The Data

import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.bar(category_high_rating.index,category_high_rating.values,color="coral",width=0.75)
plt.title("Category High Rating")
plt.xlabel("Category Name")
plt.ylabel("Average Rating")
plt.grid()
plt.show()
```

```python
#3.Is there a correlation between the product's price and its rating?

cor_product_rating=amazon["actual_price"].corr(amazon["rating"])
print(f"Correlation between the Product Price and its Rating:{cor_product_rating}")
```

```python
#4.What is the most common word in the positive and negative reviews?

from collections import Counter
import pandas as pd
import re

# Assuming you have your Amazon review data in a pandas DataFrame called df
# Replace this with the actual loading of your dataset

# Split the reviews into words (lowercase, remove non-alphabetic characters)
def clean_and_split(text):
    # Remove non-alphabetic characters and split by space
    words = re.findall(r'\b\w+\b', text.lower())  # lowercasing and splitting into words
    return words

# Filter reviews by positive (e.g., rating >= 3) and negative (rating < 3) ratings
positive_reviews = amazon[amazon['rating'] >= 3]['review_title']
negative_reviews = amazon[amazon['rating'] < 3]['review_title']

# Apply the cleaning and word-splitting function
positive_words = ' '.join(positive_reviews.dropna()).lower()
negative_words = ' '.join(negative_reviews.dropna()).lower()

positive_word_list = clean_and_split(positive_words)
negative_word_list = clean_and_split(negative_words)

# Count the frequency of each word
positive_word_counts = Counter(positive_word_list)
negative_word_counts = Counter(negative_word_list)

# Most common word in positive reviews
most_common_positive_word = positive_word_counts.most_common(1)

# Most common word in negative reviews
most_common_negative_word = negative_word_counts.most_common(1)

print("Most common word in positive reviews:", most_common_positive_word)
print("Most common word in negative reviews:", most_common_negative_word)
```
